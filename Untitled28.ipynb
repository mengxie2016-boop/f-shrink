```python
# @title 测试 f-Shrink decay=0.03, 0.04, 0.035
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# 常数 f
f = 0.4745409995126511

def f_shrink(x):
    sign = torch.sign(x)
    abs_x = torch.abs(x)
    return sign * torch.sqrt(abs_x) * (f ** (abs_x / (2 * f)))

class FShrinkScheduler:
    def __init__(self, optimizer, mode='decay', lr_init=0.1, f=0.4745409995126511, decay=0.03):
        self.optimizer = optimizer
        self.lr = lr_init
        self.f = f
        self.mode = mode
        self.decay = decay
        self._set_lr(lr_init)

    def _set_lr(self, lr):
        for param_group in self.optimizer.param_groups:
            param_group['lr'] = lr

    def step(self):
        if self.mode == 'decay':
            self.lr = self.lr * (1 - self.decay) + self.f * self.decay
        self._set_lr(self.lr)
        return self.lr

# 设置设备
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"使用设备: {device}")

# 加载 CIFAR-10
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True, num_workers=2)

# 简单 CNN
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(32 * 8 * 8, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 32 * 8 * 8)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

def train_with_decay(decay, epochs=3):
    model = SimpleCNN().to(device)
    optimizer = optim.SGD(model.parameters(), lr=0.1)
    criterion = nn.CrossEntropyLoss()
    scheduler = FShrinkScheduler(optimizer, mode='decay', decay=decay)

    losses = []
    lrs = []

    for epoch in range(epochs):
        running_loss = 0.0
        for i, data in enumerate(trainloader):
            inputs, labels = data[0].to(device), data[1].to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            scheduler.step()

            running_loss += loss.item()
            if i % 100 == 99:
                avg_loss = running_loss / 100
                losses.append(avg_loss)
                lrs.append(optimizer.param_groups[0]['lr'])
                print(f'[{epoch+1}, {i+1}] loss: {avg_loss:.4f}, lr: {lrs[-1]:.6f}')
                running_loss = 0.0

    return losses, lrs

# 训练 Cosine 作为 baseline
print("\n" + "="*60)
print("训练 Cosine...")
print("="*60)
model_cosine = SimpleCNN().to(device)
optimizer_cosine = optim.SGD(model_cosine.parameters(), lr=0.1)
criterion = nn.CrossEntropyLoss()
scheduler_cosine = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer_cosine, T_max=3 * len(trainloader))

loss_cosine = []
lr_cosine = []
for epoch in range(3):
    running_loss = 0.0
    for i, data in enumerate(trainloader):
        inputs, labels = data[0].to(device), data[1].to(device)
        optimizer_cosine.zero_grad()
        outputs = model_cosine(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer_cosine.step()
        scheduler_cosine.step()

        running_loss += loss.item()
        if i % 100 == 99:
            avg_loss = running_loss / 100
            loss_cosine.append(avg_loss)
            lr_cosine.append(optimizer_cosine.param_groups[0]['lr'])
            print(f'[{epoch+1}, {i+1}] loss: {avg_loss:.4f}, lr: {lr_cosine[-1]:.6f}')
            running_loss = 0.0

# 训练不同 decay
decays = [0.03, 0.04, 0.035]
results = {}

for decay in decays:
    print(f"\n" + "="*60)
    print(f"训练 f-Shrink decay={decay}...")
    print("="*60)
    loss, lr = train_with_decay(decay, epochs=3)
    results[decay] = {'loss': loss, 'lr': lr}

print("\n" + "="*60)
print("实验结果总结")
print("="*60)
print(f"Cosine 最终损失:      {loss_cosine[-1]:.4f}")
for decay in decays:
    print(f"f-Shrink decay={decay} 最终损失: {results[decay]['loss'][-1]:.4f}")
print(f"Cosine 最终学习率:    {lr_cosine[-1]:.6f}")
for decay in decays:
    print(f"f-Shrink decay={decay} 最终学习率: {results[decay]['lr'][-1]:.6f}")

# 画图
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(loss_cosine, label='Cosine', linewidth=2, color='black')
for decay in decays:
    plt.plot(results[decay]['loss'], label=f'f-Shrink decay={decay}', linewidth=2)
plt.xlabel('Step (per 100 batches)')
plt.ylabel('Loss')
plt.title('损失对比')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(lr_cosine, label='Cosine', linewidth=2, color='black')
for decay in decays:
    plt.plot(results[decay]['lr'], label=f'f-Shrink decay={decay}', linewidth=2)
plt.axhline(y=f, color='r', linestyle='--', label=f'f={f:.4f}')
plt.xlabel('Step (per 100 batches)')
plt.ylabel('Learning Rate')
plt.title('学习率变化')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
